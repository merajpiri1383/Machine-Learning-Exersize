# **Image Detection Jute Pest**


This dataset has 4 classes. Data are divided in three partition train, val and test. The classes are 0 : Beet Armyworm 1 : Black Hairy 2 : Cutworm 3 : Field Cricket 4 : Jute Aphid 

**Download Files** : https://archive.ics.uci.edu/dataset/920/jute+pest+dataset

### **Step 1 : import libraries**

```python
import pandas as pd
import numpy as np
import os
from skimage.io import imread
from skimage.transform import resize
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

### **Step 2 : Load Data And Preprocessing**

- convert images into numbers using imread from skimage module 
- `pip install scikit-image` 
- save images in **X_train**
- save image category in **y_train**

```python
categories = [
    "Beet Armyworm",
    "Black Hairy",
    "Cutworm",
    "Field Cricket",
]


train_dir = "train/"
test_dir = "test/"

X_train = []
y_train = []

for category in categories : 
    path = os.path.join(train_dir,category)
    items = os.listdir(path)
    for file in items : 
        file_path = os.path.join(path,file)
        img_array = imread(file_path)
        img_resized = resize(img_array,(150,150,3))
        X_train.append(img_resized.flatten())
        y_train.append(category)
    print(category)
```

```python
X_train = pd.DataFrame(X_train)
y_train = np.array(y_train)
```

### now we do the same for X_test and y_test

```python
X_test = []
y_test = []

for category in categories : 
    path = os.path.join(test_dir,category)
    items = os.listdir(path)
    for file in items : 
        file_path = os.path.join(path,file)
        img_array = imread(file_path)
        img_resized = resize(img_array,(150,150,3))
        X_test.append(img_resized.flatten())
        y_test.append(category)
    print(category)
```

```python
X_test = pd.DataFrame(X_test)
y_test = np.array(y_test)
```

### **Step 3 : Train The Model**

```python
model = SVC(gamma=10,C=10,kernel="rbf")
model.fit(X_train,y_train)
```

**i didnt use GridSearchCV because my system cant handle it**

### **Step 4 : Evaluation The model**

```python
y_predicted = model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predicted))
```

```python
print(classification_report(y_test,y_predicted))
```
